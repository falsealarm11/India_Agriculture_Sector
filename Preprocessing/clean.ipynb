```python
import pandas as pd
```

```python
dt = pd.read_csv("D:/SmartBridge/Tableau/India Agriculture Crop Production/India Agriculture Crop Production.csv")
```

```python
# Assuming your dataframe is named 'df'
crop_by_state = dt.groupby('State')['Crop'].unique().apply(list).reset_index()

crop_by_state
```

```python
crop_by_district = dt.groupby('District')['Crop'].unique().apply(list).reset_index()

crop_by_district
```

```python
most_common_crop_by_state = dt.groupby('State')['Crop'].apply(lambda x: x.mode().values[0]).reset_index()

most_common_crop_by_state
```

```python
most_common_crop_by_district = dt.groupby('District')['Crop'].apply(lambda x: x.mode().values[0]).reset_index()

most_common_crop_by_district
```

```python
dt['Crop'] = dt.groupby('District')['Crop'].transform(lambda x: x.fillna(x.mode().values[0]))
```

```python
dt.isna().sum()
```

```python
dt['Season'] = dt['Season'].fillna(dt['Season'].mode().values[0])
```

```python
dt.isna().sum()
```

```python
dt['Area'] = dt.groupby('District')['Area'].transform(lambda x: x.fillna(x.mean()))
```

```python
dt.isna().sum()
```

```python
dt['Area'] = dt['Area'].fillna(dt['Area'].mean())
```

```python
dt.isna().sum()
```

```python
dt['Production'] = dt.groupby('District')['Production'].transform(lambda x: x.fillna(x.mean()))
```

```python
dt.isna().sum()
```

```python
dt['Production'] = dt['Production'].fillna(dt['Production'].mean())
```

```python
dt.isna().sum()
```

```python
dt['Yield'] = dt.groupby('District')['Yield'].transform(lambda x: x.fillna(x.mean()))
```

```python
dt.isna().sum()
```

```python
dt['Yield'] = dt['Yield'].fillna(dt['Yield'].mean())
```
